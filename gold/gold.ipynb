```python
#importing the modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the dataset
ds=pd.read_csv('gold.csv')
```

```python
ds
```

```python
#checking for the duplicate value
ds.duplicated().sum()
```

```python
#checking for the missing value
ds.isnull().sum()
```

```python
ds.drop(['Date','Volume','Adj Close'],axis=1,inplace=True)
```

```python
#Analysing the value using the scatter plot
sns.scatterplot(ds)
```

```python
plt.plot(ds)
plt.grid()
```

```python
#opening value
plt.plot(ds['Open'])
plt.title('Opening Value')
plt.grid()
```

```python
#closing value
plt.plot(ds['Close'],color='Green')
plt.title('Closing Value')
plt.grid()
```

```python
#analysing the value along with barplot
sns.barplot(ds)
```

```python
#to check the value of the price periodically
plt.hist(ds)
```
