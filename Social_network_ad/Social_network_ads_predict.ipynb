```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
x= df.iloc[:, [2,3]].values
```

```python
x
```

```python
y=df.iloc[:, 4].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_trans=sc.fit_transform(x_train)
```

```python
x_test_trans=sc.fit_transform(x_test)
```

```python
x_train_trans,x_train_trans.shape
```

```python
x_test_trans.shape
```

```python
y_train.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,SimpleRNN
from keras.losses import MeanSquaredError
from keras.optimizers import Adam
```

```python
model=Sequential()
model.add(SimpleRNN(4,input_shape=(None,1),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='softmax'))
model.summary()
```

```python
model.compile(loss=MeanSquaredError(),optimizer=Adam(),metrics=['accuracy'])
```

```python
model.fit(x_train_trans,y_train,batch_size=55,epochs=5)
```

```python
model.evaluate(x_test_trans,y_test,batch_size=55)
```
